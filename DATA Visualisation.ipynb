# Bar Graphs

```python
lang=['C','C++','Java','Python','Php']
students=[23,17,35,29,12]
x=plt.bar(lang,students,width=0.5,color='r')
plt.show()
```

```python
#Example2
N=5
men_means=[20,35,30,35,27]
women_means=[25,32,34,20,25]
ind=np.arange(5)
width=0.35
plt.bar(ind,men_means,width,label='Men',color='r')
plt.bar(ind+width,women_means,width,label='Women',color='b')
plt.grid(True)
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title("Scores by men and women")
plt.xticks(ind,('G1','G2','G3','G4','G5'),rotation='45')
plt.legend()
plt.show()
```

```python
# example 1
labels='Python','C++','Java','C'
sizes=[215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0.1,0.1,0.1)

#plot
plt.grid(True)
plt.pie(sizes,explode=explode,colors=colors,labels=labels,autopct='%1.1f%%',pctdistance=0.4,shadow=True,startangle=150)
plt.show()
```
